```python
import pandas as pd
import numpy as np
import os
import unidecode 
```

```python
print(os.getcwd())
df_parroquias=pd.read_csv('../data_csv/codigos_estandar/parroquias/codigo_vs_parroquia.csv')
df_parroquias
```

```python
# count the number of nan values of each column
df_parroquias.isnull().sum()
```

```python
std_parroquias=pd.read_csv('../data_csv/codigos_estandar/parroquias/std_parroquias.csv')
std_parroquias
```

```python
parroquias_2023=pd.read_csv('../data_csv/seccionales/2023/diccionarios/parroquias_2023.csv')
parroquias_2023
```

```python
#match the parroquias from the std_parroquias with the parroquias from the parroquias_2023
#first we need to put the unidecode of the parroquias_2023  and std_parroquias
std_parroquias['PARROQUIA_UNIDECODE']=std_parroquias['NOMBRE_PARROQUIA'].apply(lambda x: unidecode.unidecode(x).upper())
parroquias_2023['PARROQUIA_UNIDECODE']=parroquias_2023['PARROQUIA_NOMBRE'].apply(lambda x: unidecode.unidecode(x).upper())
#also with the nombre_canton
std_parroquias['CANTON_UNIDECODE']=std_parroquias['NOMBRE_CANTON'].apply(lambda x: unidecode.unidecode(x).upper())
parroquias_2023['CANTON_UNIDECODE']=parroquias_2023['CANTON_NOMBRE'].apply(lambda x: unidecode.unidecode(x).upper())


#merge the dataframes
df_parroquias=pd.merge(std_parroquias,parroquias_2023,how='left',on=['PARROQUIA_UNIDECODE','CANTON_UNIDECODE'])
df_parroquias
```

```python
#Preserve the next columns
columns=['CODIGO_PROVINCIA','NOMBRE_PROVINCIA','CANTON_CODIGO','CODIGO_CANTON','CANTON_UNIDECODE','PARROQUIA_CODIGO','CODIGO_PARROQUIA','PARROQUIA_UNIDECODE']
df_parroquias=df_parroquias[columns]
#put it on order by PROVINCIA_CODIGO,CANTON_CODIGO,CODIGO_CANT
df_parroquias
```

```python
#Se llenaron a mano los valores faltantes
df_parroquias=pd.read_excel('../data_csv/codigos_estandar/parroquias/parroquias_2023_std.xlsx')
df_parroquias
```

```python
#put codigo_canton as a string
df_parroquias['CODIGO_CANTON']=df_parroquias['CODIGO_CANTON'].astype(str)
#put codigo_provincia as a string
df_parroquias['CODIGO_PROVINCIA']=df_parroquias['CODIGO_PROVINCIA'].astype(str)
#put codigo_parroquia as a string
df_parroquias['CODIGO_PARROQUIA']=df_parroquias['CODIGO_PARROQUIA'].astype(str)


# in the canton_codigo column if the value has 3 digits we need to put a 0 in front of the number

for i in range(len(df_parroquias)):
    #print (len(str(df_parroquias.loc[i,'CODIGO_CANTON'])))
    if len(str(df_parroquias.loc[i,'CODIGO_CANTON']))==5:
        #print (df_parroquias.loc[i,'CODIGO_CANTON'])
        #print (df_parroquias.loc[i,'CODIGO_CANTON'])
        df_parroquias.loc[i,'CODIGO_CANTON']='0'+str(df_parroquias.loc[i,'CODIGO_CANTON'])

# in the provincia_codigo column if the value has 2 digits we need to put a 0 in front of the number
for i in range(len(df_parroquias)):
    #print (len(str(df_parroquias.loc[i,'CODIGO_CANTON'])))
    if len(str(df_parroquias.loc[i,'CODIGO_PROVINCIA']))==3:
        #print (df_parroquias.loc[i,'CODIGO_CANTON'])
        #print (df_parroquias.loc[i,'CODIGO_CANTON'])
        df_parroquias.loc[i,'CODIGO_PROVINCIA']='0'+str(df_parroquias.loc[i,'CODIGO_PROVINCIA'])
        
# in the parroquia_codigo column if the value has 2 digits we need to put a 0 in front of the number
for i in range(len(df_parroquias)):
    #print (len(str(df_parroquias.loc[i,'CODIGO_CANTON'])))
    if len(str(df_parroquias.loc[i,'CODIGO_PARROQUIA']))==7:
        #print (df_parroquias.loc[i,'CODIGO_CANTON'])
        #print (df_parroquias.loc[i,'CODIGO_CANTON'])
        df_parroquias.loc[i,'CODIGO_PARROQUIA']='0'+str(df_parroquias.loc[i,'CODIGO_PARROQUIA'])        
df_parroquias
```

```python
#add a "C" in front of the codigo_canton
df_parroquias['CODIGO_CANTON']='C'+df_parroquias['CODIGO_CANTON']

#add a "P" in front of the codigo_provincia
df_parroquias['CODIGO_PROVINCIA']='P'+df_parroquias['CODIGO_PROVINCIA']

#add a "PR" in front of the codigo_parroquia if not nan
df_parroquias['CODIGO_PARROQUIA']=df_parroquias['CODIGO_PARROQUIA'].apply(lambda x: 'Q'+str(x) if pd.notnull(x) else x)

df_parroquias
```

```python
df_parroquias=pd.read_excel('../data_csv/codigos_estandar/parroquias/std_parroquias_v2.xlsx')
df_parroquias
```

```python
std_cantones=pd.read_csv('../data_csv/codigos_estandar/cantones/std_cantones.csv')
std_cantones
```

```python

#mergue the dataframes
#in std cantones is CANTON_CODIGO and in std_parroquias is CODIGO_CANTON
df_parroquias_std=pd.merge(df_parroquias,std_cantones,how='left',left_on='CODIGO_CANTON',right_on='CANTON_CODIGO')

df_parroquias_std
```

```python
#Print the rows that are in df_parroquias_std and not in df_parroquias
df_parroquias_std[~df_parroquias_std['CANTON_NOMBRE'].isin(df_parroquias['CANTON_UNIDECODE'])]
```

```python
# Preserve the next columns
columns=['PROVINCIA_CODIGO','NOMBRE_PROVINCIA','CANTON_CODIGO_y','CANTON_NOMBRE','PARROQUIA_CODIGO','CODIGO_PARROQUIA','PARROQUIA_UNIDECODE']
df_parroquias_std=df_parroquias_std[columns]
df_parroquias_std
```

```python
#rename the columns
df_parroquias_std.rename(columns={'CANTON_CODIGO_y':'CODIGO_CANTON'},inplace=True)
df_parroquias_std.rename(columns={'PARROQUIA_CODIGO':'PARROQUIA_CODIGO_OLD'},inplace=True)
df_parroquias_std.rename(columns={'CODIGO_PARROQUIA':'PARROQUIA_CODIGO'},inplace=True)
df_parroquias_std.rename(columns={'PARROQUIA_UNIDECODE':'PARROQUIA_NOMBRE'},inplace=True)

df_parroquias_std
```

```python
# Revisar qué parroquias tienen código "Qnan"
df_parroquias_std[df_parroquias_std['PARROQUIA_CODIGO']=='Qnan']
```

```python
#Colocar parroquia_codigo_old como int
df_parroquias_std['PARROQUIA_CODIGO_OLD']=df_parroquias_std['PARROQUIA_CODIGO_OLD'].apply(lambda x: int(x) if pd.notnull(x) else x)
df_parroquias_std
```

```python
#Para las parroquias que tienen código "Qnan" se les asigna el código de PARROQUIA_CODIGO_OLD con la Q adelante
df_parroquias_std['PARROQUIA_CODIGO']=df_parroquias_std.apply(lambda x: 'Q'+str(x['PARROQUIA_CODIGO_OLD']) if x['PARROQUIA_CODIGO']=='Qnan' else x['PARROQUIA_CODIGO'],axis=1)
df_parroquias_std
```

```python
# Revisar qué parroquias tienen código "Qnan"
df_parroquias_std[df_parroquias_std['PARROQUIA_CODIGO']=='Qnan']
```

```python
#revisar que no hayan valores repetidos
df_parroquias_std[df_parroquias_std.duplicated(subset=['PARROQUIA_CODIGO'],keep=False)]
```

```python
#hacer un sort por provincia_nombre,canton_nombre,parroquia_nombre
df_parroquias_std.sort_values(by=['PROVINCIA_CODIGO','CANTON_NOMBRE','PARROQUIA_NOMBRE'],inplace=True)
df_parroquias_std
```
