```python
import pandas as pd
import numpy as np
import os
import unidecode
```

```python
input_folder = '../data_csv'
file_name = 'resultados'
def recolect_column_names(input_folder, file_name):
    df_info = pd.DataFrame()
    max_length = 0  # Variable to store the maximum length of columns among all files
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.endswith(".csv"):
                if file.startswith(file_name):
                    file_path = os.path.join(root, file)
                    df = pd.read_csv(file_path, low_memory=False)
                    df.columns = [unidecode.unidecode(col) for col in df.columns]
                    anio = file.split(file_name)[1].split(".")[0][1:]  # Extracting year from filename
                    #print(anio)
                    df_info = pd.concat([df_info, pd.DataFrame({'Columnas': df.columns, 'Año': anio})])
                    if len(df.columns) > max_length:
                        max_length = len(df.columns)
    df_info = df_info.reset_index(drop=True)
    return df_info


columnas_result=recolect_column_names(input_folder, file_name)
                    
```

```python
print(columnas_result)
```

```python
def crear_df_info(df_info, column_to_study, column_to_use):
    df_info_std = pd.DataFrame()
    # Iterar sobre las filas del DataFrame
    for index, row in df_info.iterrows():
        # Print something to check that it is running
        print(row["Año"])
        # Revisar si la columna que se quiere recuperar el código ya está en el dataframe
        if row[column_to_study] not in df_info_std.index:
            # Si la columna que se quiere recuperar el código no está en el dataframe, agregarla y colocar su código en la columna correspondiente al año de la elección
            df_info_std.loc[row[column_to_study], row["Año"]] = row[column_to_use]
        else:
            # Si la columna que se quiere recuperar el código ya está en el dataframe, colocar su código en la columna correspondiente al año de la elección
            df_info_std.loc[row[column_to_study], row["Año"]] = row[column_to_use]
    
    # Colocar las columnas de los años en orden
    df_info_std = df_info_std.reindex(sorted(df_info_std.columns), axis=1)
    # Colocar las filas en orden
    df_info_std = df_info_std.sort_index()
    return df_info_std

df_resultados = crear_df_info(columnas_result, "Columnas", "Año")
```

```python
df_resultados.to_csv("../src/open_elec/data/Codigos_estandar/resultados/df_resultados.csv")
print(df_resultados)
```

```python
# Ahora lo mismo con el registro_electoral
file_name = 'registro_electoral'
columnas_registro=recolect_column_names(input_folder, file_name)
df_registro = crear_df_info(columnas_registro, "Columnas", "Año")
print(df_registro)
df_registro.to_csv("../src/open_elec/data/Codigos_estandar/registro_electoral/df_registro.csv")
```
